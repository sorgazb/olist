```python
import numpy as np
import pandas as pd
```

```python
df_order_reviews = pd.read_csv("./olist_order_reviews_dataset.csv", encoding='utf-8')
df_costumers = pd.read_csv("./olist_customers_dataset.csv", encoding='utf-8')
df_orders_items = pd.read_csv("./olist_order_items_dataset.csv", encoding='utf-8')
df_order_payments = pd.read_csv("./olist_order_payments_dataset.csv", encoding='utf-8')
df_order = pd.read_csv("./olist_orders_dataset.csv", encoding='utf-8')
df_product_category = pd.read_csv("./olist_product_category_name_translation.csv", encoding='utf-8')
df_product = pd.read_csv("./olist_products_dataset.csv", encoding='utf-8')
df_seller = pd.read_csv("./olist_sellers_dataset.csv", encoding='utf-8')
```

```python

# df_order_reviews.groupby('review_score').nunique()

#cambiar a datetime
df_order_reviews['review_creation_date'] = pd.to_datetime(df_order_reviews['review_creation_date'])
df_order_reviews['review_answer_timestamp'] = pd.to_datetime(df_order_reviews['review_answer_timestamp'])

ranking = {
    1: 'pesimo',
    2: 'bajo',
    3: 'bien',
    4: 'recomendo',
    5: 'super recomendo'
}

#rellenar nulos
df_order_reviews['review_comment_title'] = df_order_reviews['review_comment_title'].fillna(df_order_reviews['review_score'].map(ranking))
df_order_reviews['review_comment_message'] = df_order_reviews['review_comment_message'].fillna('No comment')

df_order_reviews
```

```python
#dar formato datetime
df_order['order_purchase_timestamp'] = pd.to_datetime(df_order['order_purchase_timestamp'])
df_order['order_approved_at'] = pd.to_datetime(df_order['order_approved_at'])
df_order['order_delivered_carrier_date'] = pd.to_datetime(df_order['order_delivered_carrier_date'])
df_order['order_delivered_customer_date'] = pd.to_datetime(df_order['order_delivered_customer_date'])
df_order['order_estimated_delivery_date'] = pd.to_datetime(df_order['order_estimated_delivery_date'])
```

```python

#cambiar idioma categoria
name = df_product_category.set_index('product_category_name')['product_category_name_english'].to_dict()
df_product['product_category_name'] = df_product['product_category_name'].map(name)
df_product
```

```python
#rellenar nulos 
df_product['product_category_name'] = df_product['product_category_name'].fillna('Not specified')
df_product['product_name_lenght'] = df_product['product_name_lenght'].fillna(df_product['product_name_lenght'].mean()).round()
df_product['product_description_lenght'] = df_product['product_description_lenght'].fillna(df_product['product_description_lenght'].mean()).round()
df_product['product_photos_qty'] = df_product['product_photos_qty'].fillna(df_product['product_photos_qty'].mean()).round()
df_product['product_weight_g'] = df_product['product_weight_g'].fillna(df_product['product_weight_g'].mean()).round()
df_product['product_length_cm'] = df_product['product_length_cm'].fillna(df_product['product_length_cm'].mean()).round()
df_product['product_height_cm'] = df_product['product_height_cm'].fillna(df_product['product_height_cm'].mean()).round()
df_product['product_width_cm'] = df_product['product_width_cm'].fillna(df_product['product_width_cm'].mean()).round()

df_product
```
